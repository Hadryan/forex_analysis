```python
import pandas as pd
import numpy as np
import re
from matplotlib import pyplot as plt
```

```python
with open('../data/XXBTZEUR.csv') as f:
    # Reading header of numpy file
    header = f.readline()[2:-1]
    # Converting it from str to a list
    header = [s.strip() for s in re.split(",\s", header)]
df_btc = pd.DataFrame(data=np.loadtxt('../data/XXBTZEUR.csv'), columns=header) 
df_btc
```

```python
with open('../data/forex.csv') as f:
    # Reading header of numpy file
    header = f.readline()[2:-1]
    # Converting it from str to a list
    header = [s.strip() for s in re.split(",\s", header)]
df_forex = pd.DataFrame(data=np.loadtxt('../data/forex.csv'), columns=header) 
df_forex
```
